```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#test from jupy
np.random.seed(1)
# Regression Example
# We try to predict someone's height (y) based off of their arm length (x)

x = np.array([30.17, 49.04, 29.21, 42.96, 42.01, 33.06, 37.99, 47.42, 50.46, 28.32, 31.47, 33.73, 35.69, 37.78, 56.98])
y = np.array([116.93, 192.81, 114.38, 169.86, 165.90, 128.94, 147.92, 186.74, 198.38, 110.65, 122.50, 131.71, 139.68, 148.13, 224.47])

# I made the error too small when making the numbers, I'm adding some more error
y = y + np.random.normal(loc = 0, scale = 3, size = np.size(y))


```

```python
ones = np.array([1] * np.size(x))

X = np.column_stack((ones, x)) # X is a matrix with a column of 1's and a column with 'x'

# Linear regression is done by solving for "b" in the following matrix equation
#    X'Xb = X'y
# This is done in python using b = np.linalg.solve(X'X, X'y)
b = np.linalg.solve(np.dot(np.transpose(X), X), np.dot(np.transpose(X), y))

b

# Then estimates from the model are given by y = Xb

# We estimate that y = -4.9455257 + 4.04195271*x
# This is pretty close to how I actually generated y
# which was y = -3 + 4*x
```

```python

plt.scatter(x, y)

plt.scatter(x,b[0] + b[1]*x,color='green') # all of the green dots are our predicted values based on our model
# but our model can actually predict height (y) based off ANY given arm length (x) NOT just the ones from our data
plt.plot(x,b[0] + b[1]*x,color='black')
```
